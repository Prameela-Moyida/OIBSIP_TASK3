```python
x=df.drop(columns=['Species'])
y=df['Species']
```

```python
print(x)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
X_train
```

```python
 y_test
```

```python
#Supprot Vector Machine
from sklearn.svm import SVC
from sklearn.svm import SVC
```

```python
model_svc=SVC()
model_svc.fit(X_train,y_train)
```

```python
pre1=model_svc.predict(X_test)
#calculate the Accuracy
from sklearn.metrics import accuracy_score

print("Support vector Machine Accuracy:",accuracy_score(y_test, pre1)*100)
```

```python
#Logistic  Regression
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(X_train,y_train)
```

```python
pre2=LR.predict(X_test)
#Calculate Accuracy
from sklearn.metrics import accuracy_score
print("Logistic Regression Accuracy:",accuracy_score(y_test,pre2)*100)
```

```python
model_svc.predict([[6.0,4.2,5.0,6.0]])
```
